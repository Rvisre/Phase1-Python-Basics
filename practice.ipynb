Using a for loop to iterate through Data Frames

```python
for index , row in coffee.iterrows():
    print(index)
    print(row)
    print("\n\n\n\n\n\n")
```

Iterating through a specific row

```python
for index , row in coffee.iterrows():
    print(index)
    print(row["Units Sold"])
    print("\n\n\n\n\n\n")
```
